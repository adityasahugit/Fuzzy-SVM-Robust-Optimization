```python
import pandas as pd
import numpy as np
from sklearn import cross_validation
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("Australian.csv")
train.head()
```

```python
train['0.3'] = train['0.3'].map({1: 1, 0: -1})
print(train['0.3'])
```

```python
features = train.columns[0:13]
X = train[features]
y = train['0.3']
train.head()
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size=0.2,random_state=0)
C= 0.1
```

```python
clf_svm = svm.SVC(kernel='linear', C=C)
clf_svm.fit(X_train, y_train)
y_pred_svm = clf_svm.predict(X_test) 
acc_svm = accuracy_score(y_test, y_pred_svm)
print ("Overall Linear SVM accuracy: ",acc_svm)
```

```python
clf_svm = svm.SVC(kernel='rbf', gamma=0.0001, C=C)
clf_svm.fit(X_train, y_train)
y_pred_svm = clf_svm.predict(X_test) 
acc_svm = accuracy_score(y_test, y_pred_svm)
print ("Overall RBF KERNEL SVM accuracy: ",acc_svm)
```
